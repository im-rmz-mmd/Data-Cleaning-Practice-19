```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('auto-mpg.csv')
df
```

```python
df = df.drop(columns = ['car name'])
df = df.drop_duplicates()
df = df.dropna()
df
```

```python
df = df.rename({'model year' : 'model_year'})
df
```

```python
m1 = df['mpg'].quantile(0.25)
m2 = df['mpg'].quantile(0.75)

m1, m2
```

```python
a = m2 - m1

lower_bound = m1 - 1.5 * a
upper_bound = m2 - 1.5 * a

lower_bound, upper_bound
```

```python
x = df[(df['mpg'] >= lower_bound) & (df['mpg'] <= upper_bound)]
x
```

```python
df = df.drop([25, 26, 27, 28, 67, 103, 124])
df
```

```python
w1 = df['weight'].quantile(0.25)
w2 = df['weight'].quantile(0.75)

w1, w2
```

```python
b = w2 - w1

lower_bound = w1 - 1.5 * b
upper_bound = w2 - 1.5 * b

lower_bound, upper_bound
```

```python
y = df[(df['weight'] >= lower_bound) & (df['weight'] <= upper_bound)]
y
```
